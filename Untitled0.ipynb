```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Load sample ECG data
data = pd.read_csv('sample_ecg.csv')  # Column: 'ECG'
ecg_signal = data['ECG']

# Detect R-peaks
peaks, _ = find_peaks(ecg_signal, distance=50)  # distance can be adjusted
heart_rate = len(peaks) * (60 / (len(ecg_signal)/250))  # assuming 250 Hz sampling

print(f"Estimated Heart Rate: {heart_rate:.2f} bpm")

# Plot ECG signal and detected peaks
plt.figure(figsize=(10,4))
plt.plot(ecg_signal, label='ECG Signal')
plt.plot(peaks, ecg_signal[peaks], "x", label='Detected Peaks')
plt.title("ECG Signal with R-Peaks")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

```python
# Create a sample_ecg.csv file
import numpy as np

# Generate a synthetic ECG-like signal
fs = 250  # Sampling frequency
t = np.arange(0, 10, 1/fs) # 10 seconds of data
ecg_signal_dummy = 0.5 * np.sin(2 * np.pi * 1 * t) + 0.1 * np.sin(2 * np.pi * 10 * t) # Basic sine waves

# Add some noise
noise = np.random.normal(0, 0.05, len(t))
ecg_signal_dummy = ecg_signal_dummy + noise

# Simulate R-peaks
r_peaks_indices = np.arange(0, len(t), fs * 0.8) # Approximately 75 bpm
for i in r_peaks_indices:
    if int(i) < len(ecg_signal_dummy):
        ecg_signal_dummy[int(i)] += 1.5 # Add a peak

# Save to a CSV file
dummy_data = pd.DataFrame({'ECG': ecg_signal_dummy})
dummy_data.to_csv('sample_ecg.csv', index=False)

print("Created  'sample_ecg.csv' file.")
```
